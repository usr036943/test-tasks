## Задание № 1 (Numpy):

1. Создать двумерный массив из случайных чисел от 0 до 1 размером 10х10 и вывести на экран.
2. Найти индекс (номер строки и номер столбца) максимального элемента.
3. Переставить столбцы так, чтобы в первой строке массива значения шли по возрастанию.
4. Отнормировать значения в каждой строке (то есть для каждой строки найти сумму элементов
и разделить на эту сумму каждый элемент строки).
5. Заменить значения массива, которые больше 0.5, на -1.
6. Вывести полученный массив на экран.

## Решение: 

### 1. Создать двумерный массив из случайных чисел от 0 до 1 размером 10х10 и вывести на экран.

```python
# Импортируем необходимые библиотеки

import pandas as pd
import numpy as np
import random
```

```python
rand_arr = np.random.rand(10, 10) # cоздаем двумерный массив из случайных чисел 
rand_arr # выведем на экран массив 
```

### 2. Найти индекс (номер строки и номер столбца) максимального элемента.

```python
max_row, max_col = np.unravel_index(rand_arr.argmax(), rand_arr.shape) # найдем индекс маскимального элемента 
print("Максимальный элемент находится в строке", max_row, "и в столбце", max_col) # выведем на экран строки и столбца
```

### 3. Переставить столбцы так, чтобы в первой строке массива значения шли по возрастанию.

```python
rand_arr = rand_arr[::, rand_arr[0,].argsort()[::1]] # отсортируем массив с помощью функции argsort() 
```

```python
rand_arr # проверим сортировку 
```

### 4. Отнормировать значения в каждой строке (то есть для каждой строки найти сумму элементов и разделить на эту сумму каждый элемент строки).

```python
rand_arr_normalized = rand_arr/rand_arr.sum(axis=1) # делим каждый элемент на сумму элементов 
```

```python
rand_arr_normalized # выведем отнормированный массив на экран 
```

### 5. Заменить значения массива, которые больше 0.5, на -1.

Для наглядности заменим значения в исходном массиве, т.к. в отнормированном массиве все значения меньше 0,5. 

```python
rand_arr[rand_arr>0.5] = -1 # заменим значения массива, которые больше 0.5, на -1
```

### 6. Вывести полученный массив на экран.

```python
rand_arr # выведем массив на экран 
```
